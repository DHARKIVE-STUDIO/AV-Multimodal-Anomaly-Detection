```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def show_camera_image(camera_image, camera_labels, layout, cmap = None):

#Display the camerage image and the given camera labels
  ax = plt.subplot(*layout)
  for camera_labels in frame.camera_labels:

    if camera_labels.name != camera_image.name:
      continue

    for label in camera_labels.labels:
      #Bounding box
      ax.add_patch(patches.Rectangle(
          xy = (label.box.center_x-0.5*label.box.length,
                label.box.center_y-0.5*label.box.width),
          width=label.box.length,
          height=label.box.width,
          linewidth=1,
          edgecolor='red',
          facecolor='none'))

    #Display the camera image
  plt.imshow(tf.image.decode_jpeg(camera_image.image), cmap=cmap)
  plt.title(open_dataset.CameraName.Name.Name(camera_image.name), fontsize=20)
  plt.grid(False)
  plt.axis('off')

plt.figure(figsize=(25, 20))

for index, image in enumerate(frame.images):
  show_camera_image(image, frame.camera_labels, [3,3, index+1])
```

```python
import tensorflow as tf
from waymo_open_dataset import dataset_pb2 as open_dataset
import matplotlib.pyplot as plt
import numpy as np

tfrecord_file_path = '/content/drive/MyDrive/IS/testing_3d_camera_only_detection/segment-17112899600450701754_2464_174_2484_174.tfrecord'

#Create empty Waymo frame and Parse Waymo TRFRecord
def parse_waymo_tfrecord(record):

  frame = open_dataset.Frame()
  frame.ParseFromString(record)
  return frame

dataset = tf.data.TFRecordDataset(tfrecord_file_path)

#Taken only the first frame for example purpose
for raw_record in dataset.take(1):
  frame = parse_waymo_tfrecord(raw_record.numpy())
  print(f"\nNumber of images in this frame: {len(frame.images)}")

  #Access and print the details about the images
  for i, image in enumerate(frame.images):
    print(f"\nImage {i} : camera {image.name}")

    try:
      #Decode the JPEG encoded image
      image_decoded = tf.image.decode_jpeg(image.image)

      #Convert TensorFlow to a Numpy array for visualization
      image_np = image_decoded.numpy()

      #Dispplaying the image using matplotlib
      plt.imshow(image_np)
      plt.title(f"Camera {image.name}")
      plt.axis("off")
      plt.show()
    except Exception as e:
      print(f"Error decoding or displaying image {i} : {e}")

  #Print the details about LiDar Data
  for i, laser in enumerate(frame.lasers):
    print(f"Laser {i}: {len(laser.ri_return1.range_image_compressed)} points compressed.")
```

```python
import os
import tensorflow as tf
from waymo_open_dataset import dataset_pb2 as open_dataset
import matplotlib.pyplot as plt
import numpy as np
import torch

#Loading the YOLOV5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

tfrecord_dir = "/content/drive/MyDrive/IS/testing_3d_camera_only_detection/"

#Get list of all the TFRecord files in the directory
tfrecord_all_files  =  [os.path.join(tfrecord_dir,f) for f in os.listdir(tfrecord_dir) if f.endswith('.tfrecord')]

def parse_waymo_tfrecord(record):
  frame = open_dataset.Frame()
  frame.ParseFromString(record)
  return frame

#Iterate through each TFRecord file
for tfrecord_file_path in tfrecord_all_files:
  print(f"Processing file: {tfrecord_file_path}")

  dataset = tf.data.TFRecordDataset(tfrecord_file_path)

  for raw_record in dataset:
    frame = parse_waymo_tfrecord(raw_record.numpy())

    for i, image in enumerate(frame.images):
      print(f"Processing image from camera {image.name}")

      #Decode the image using the tensorflow
      try:
          image_decoded = tf.image.decode_jpeg(image.image)
          image_np = image_decoded.numpy()

          #Here we run the YOLOv5 objection detection model on image
          results = model(image_np)

          #Display the image detection results
          results.show()
          results.save(f'/content/drive/MyDrive/results/image_{image.name}_{i}.jpg')

      except Exception as e:
          print(f"Error decoding or processing image {i} from camera {image.name} : {e}")
```
